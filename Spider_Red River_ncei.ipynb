```python
import os
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup

def DownLoadFiles(year, url):
    #url = 'https://www.ncei.noaa.gov/data/global-summary-of-the-day/access/1929/'
    html = urlopen(url) #获取网页
    bs = BeautifulSoup(html, 'html.parser') #解析网页
    hyperlink = bs.find_all('a')  #获取所有超链接
    i = 0
    for h in hyperlink:
        hh = h.get('href')
        hhh = url + hh
        print(hhh)

        file = 'D:/workspace/jupyterNotebook/张继辉红河数据爬虫代码/files/' + str(year) + '/'
        if not os.path.exists(file):
            os.makedirs(file)
        
        if i > 4:
            urll = hhh
            
            file = file + hh
            r = requests.get(urll)
            with open(file, "wb") as code:
                code.write(r.content)

        i = i + 1

if __name__ == '__main__':
    #已经爬到1944年了，下次从1945年开始爬
    for i in range(2001, 2022):
        url = 'https://www.ncei.noaa.gov/data/global-summary-of-the-day/access/' + str(i) + '/'
        print(url)
        DownLoadFiles(i, url)
```

```python
#改进版，可以从某个文件开始爬
import os
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup

def DownLoadFiles(year, no, url):
    #url = 'https://www.ncei.noaa.gov/data/global-summary-of-the-day/access/1929/'
    html = urlopen(url) #获取网页
    print("正在解析网页")
    bs = BeautifulSoup(html, 'html.parser') #解析网页
    hyperlink = bs.find_all('a')  #获取所有超链接
    i = 0
    hlen = str(len(hyperlink))
    print("总链接数：" + str(hlen))
    print("定位至：第 " + str(no) + " 条")
    for h in hyperlink:
        if (i < no):
            i = i + 1
            continue
        hh = h.get('href')
        hhh = url + hh
        print(str(i) + "/" + hlen + "  ", end="")
        print(hhh, end="")

        file = 'D:/workspace/jupyterNotebook/张继辉红河数据爬虫代码/files/' + str(year) + '/'
        if not os.path.exists(file):
            os.makedirs(file)
        
        if i > 4:
            urll = hhh
            
            file = file + hh
            r = requests.get(urll)
            with open(file, "wb") as code:
                code.write(r.content)
                print("下载完成")

        i = i + 1

if __name__ == '__main__':
    #已经爬到1944年了，下次从1945年开始爬
    for i in range(2021, 2022):
        url = 'https://www.ncei.noaa.gov/data/global-summary-of-the-day/access/' + str(i) + '/'
        print(url)
        no = 12126  #默认从第0条开始
        DownLoadFiles(i, no, url)
```
